# Analisis de principios de normalizacion

##### Análisis de Violaciones a la Primera Forma Normal (1NF)

Se evaluan dos puntos críticos:
1. **Atomicidad**: Verificar si hay columnas que contengan listas o múltiples valores en una sola celda.
2. **Valores Duplicados**: Verificar si existen registros idénticos que rompan la unicidad de las filas.

````python
import pandas as pd

# 1. Verificar duplicados exactos
duplicados = ordersFull_df.duplicated().sum()
print(f"Número de filas duplicadas exactas: {duplicados}")

# 2. Verificar atomicidad buscando separadores en cadenas
print("\nVerificando atomicidad (buscando separadores como ',' ';' '|' en strings):")
separadores = [',', ';', '\\|'] # Lista de posibles separadores

for col in ordersFull_df.columns:
    # Solo se analizan columnas de tipo objeto (strings)
    if ordersFull_df[col].dtype == 'object':
        # Se valida si algún valor contiene los separadores (ignorando nulos)
        contains_separator = ordersFull_df[col].astype(str).str.contains('|'.join(separadores), regex=True).any()

        if contains_separator:
            print(f"- Alerta: La columna '{col}' podría contener valores no atómicos (se detectaron separadores o colecciones).")
        else:
            print(f"- Columna '{col}': OK (Atómica)")
    else:
        print(f"- Columna '{col}': OK (Numérica/Fecha)")

# 3. Análisis de granularidad
print(f"\nRegistros totales: {len(ordersFull_df)}")
print(f"IDs de orden únicos: {ordersFull_df['order_id'].nunique()}")

if len(ordersFull_df) > ordersFull_df['order_id'].nunique():
    print("\nResultado 1NF: Aunque los valores individuales parecen atómicos, la estructura denormalizada repite el 'order_id'. Para cumplir estrictamente con 1NF, cada fila debe ser única y la clave primaria (ej. order_id + product_id) debe identificar unívocamente cada registro.")
````

##### Demostración de Duplicidad en `order_id`

Se identifican los `order_id` que aparecen más de una vez debido a la denormalización:

````python
# Contar las ocurrencias de cada order_id
order_counts = ordersFull_df['order_id'].value_counts()

# Filtrar solo aquellos que se repiten
repeated_orders = order_counts[order_counts > 1]

print(f"Total de órdenes con múltiples registros: {len(repeated_orders)}")
print("\nEjemplos de order_id repetidos y su frecuencia:")
display(repeated_orders.head(5))

# Mostrar el detalle de una orden específica para ver por qué se repite
example_id = repeated_orders.index[0]
print(f"\nDetalle completo para la orden {example_id}:")
display(ordersFull_df[ordersFull_df['order_id'] == example_id])
````

##### Análisis de Violaciones a la Segunda Forma Normal (2NF)

La 2NF establece que no deben existir **dependencias parciales**. En una tabla denormalizada como `ordersFull_df`:
1. La clave primaria es compuesta (ej. `order_id` + `product_id`).
2. Atributos como `customer_id`, `order_status` y `order_purchase_timestamp` dependen **solo** de `order_id`, no del `product_id`.

Esto confirma que para llegar a 2NF, debemos separar la información en tablas independientes donde los atributos dependan totalmente de su clave.

````python
# Identificar dependencias parciales
# Se toma una orden que tenga múltiples productos para demostrar la redundancia

# Se buscan órdenes con más de un ítem
order_item_counts = ordersFull_df.groupby('order_id')['order_item_id'].nunique()
multiple_items_orders = order_item_counts[order_item_counts > 1].index

if len(multiple_items_orders) > 0:
    example_2nf = multiple_items_orders[0]
    cols_order = ['order_id', 'customer_id', 'order_status', 'product_id', 'price']

    print(f"Análisis de dependencia parcial para la orden: {example_2nf}")
    display(ordersFull_df[ordersFull_df['order_id'] == example_2nf][cols_order])

    print("\nObservación:")
    print("- 'customer_id' y 'order_status' se repiten para cada producto.")
    print("- Estos atributos dependen SOLO de 'order_id'.")
    print("- Esto es una VIOLACIÓN de la 2NF en esta estructura unificada.")
else:
    print("No se encontraron órdenes con múltiples ítems para la demostración.")
````

###### Análisis de Violaciones a la Tercera Forma Normal (3NF)

La 3NF prohíbe las **dependencias transitivas**. Una dependencia transitiva ocurre cuando un atributo no clave depende de otro atributo no clave, en lugar de depender directamente de la clave primaria.

**Ejemplo teórico:** Si tuviéramos información del vendedor en esta tabla, el `vendedor_ciudad` dependería del `seller_id`, y el `seller_id` dependería de la orden. Eso es una dependencia transitiva.

````python
# Ejemplo de Dependencia Transitiva en el contexto de Reviews
# review_score y review_comment dependen de review_id, no directamente de order_id

# Se busca una orden que tenga una reseña
cols_3nf = ['order_id', 'review_id', 'review_score', 'review_creation_date']
example_3nf_df = ordersFull_df[ordersFull_df['review_id'].notnull()][cols_3nf].head(3)

print("Análisis de dependencia transitiva (Reviews):")
display(example_3nf_df)

print("\nObservación:")
print("- 'review_score' y 'review_creation_date' son atributos de la RESEÑA (review_id).")
print("- Aunque están asociados a 'order_id', su dependencia funcional directa es con 'review_id'.")
print("- En un modelo 3NF, la información de la reseña debería vivir en su propia tabla para evitar anomalías de actualización.")
````

# Trade-Offs

Ventajas de alta normalizacion:

*   Aumenta la consistencia de los datos,
*   Hay baja redundancia y por se usa menos espacio en disco
*   Las actualizaciones de informacion se hacen en menores

Desventajas de alta normalizacion:

*  Un alto grado de normalizacion penaliza el rendimiento de las consultas
*  A partir de la 4FN la flexibilidad para realizar ajustes al esquema disminuyen
*  La complejidad de las consultas aumenta

# Casos para denormalizacion

*  Patron de acceso: Heavy-Reads. Es cuando las lecturas son mas frecuentes que
   las escrituras. Ejemplos: reportes, analisis de datos, tableros.
*  Transaccionalidad baja: los datos cambian con poca frecuencia o son inmutable: catalogos de productos
*  Microservicios/Dominios: cuando existen varias bases de datos que se consultan en paralelo para hacer composiciones.

````python
def check_cardinality(df_a_name, df_b_name, key):
    if df_a_name not in dfs or df_b_name not in dfs:
        print(f"⚠️ Error: No se pudo encontrar '{df_a_name}' o '{df_b_name}' en dfs.")
        return

    df_a = dfs[df_a_name]
    df_b = dfs[df_b_name]

    # Verificar si la columna existe en ambos dataframes
    if key not in df_a.columns or key not in df_b.columns:
        print(f"❌ Error: La columna '{key}' no existe en {df_a_name} o {df_b_name}.")
        return

    count_a = df_a[key].value_counts().max()
    count_b = df_b[key].value_counts().max()

    card_a = "1" if count_a == 1 else "N"
    card_b = "1" if count_b == 1 else "N"

    print(f"Relación [{df_a_name}] ({card_a}) <---> ({card_b}) [{df_b_name}] vía '{key}'")

print("--- Análisis de Cardinalidades ---\n")

# Relaciones desde Orders
check_cardinality('orders', 'order_items', 'order_id')
check_cardinality('orders', 'order_payments', 'order_id')
check_cardinality('orders', 'order_reviews', 'order_id')
check_cardinality('customers', 'orders', 'customer_id')
check_cardinality('products', 'order_items', 'product_id')
check_cardinality('sellers', 'order_items', 'seller_id')
check_cardinality('product_category_name_translation', 'products', 'product_category_name')
````

### Diseño de Esquema Flexible y Patrones Aplicado

Para la colección products de Ecommify, se diseñó un esquema híbrido y flexible que responde a la naturaleza dinámica del catálogo:

1.  Atributos comunes: Se definieron en la raíz del documento los campos obligatorios para el negocio: _id, name, category y price.

2. Esquema flexible (Polimorfismo): Se implementó un subdocumento llamado specifications estructurado como un mapa clave-valor. Esto permite que diferentes categorías de productos almacenen atributos variables (ej. longitud del nombre, cantidad de fotos) sin forzar un esquema rígido o campos nulos masivos.

3. Computed Pattern: Se aplicó este patrón calculando previamente e insertando las métricas total_units_sold y average_rating directamente en el documento del producto.

**Decisiones Arquitectónicas: Ventajas y Trade-offs **

1. Estrategia Referencing vs Embedding

Decisión: Se optó por Referencing (referencias cruzadas) para las colecciones de reviews y sellers.
Ventaja: Evita el anti-patrón de arreglos infinitos (Unbound Arrays). Si un producto exitoso tiene miles de reseñas, incrustarlas (Embedding) superaría el límite de 16MB por documento de MongoDB y degradaría el rendimiento de lectura del catálogo.
Trade-off: Requiere realizar operaciones adicionales (como $lookup o múltiples consultas desde el backend) cuando se necesita ver el detalle de los comentarios de una reseña o los datos del vendedor.

2. Uso del Computed Pattern

Ventaja: Optimiza drásticamente las lecturas. La vista principal del catálogo Ecommify puede mostrar los productos más vendidos y mejor calificados con una sola consulta rápida, sin tener que calcular promedios en tiempo real cruzando miles de registros.
Trade-off: Añade complejidad a las escrituras. Cada vez que se concreta una nueva venta o entra una nueva reseña, se debe disparar un proceso para actualizar los contadores en el documento del producto.


### CALCULO DE METRICAS, ARMADO Y INSERCION A BASE DE DATOS 01_catalogo_enriquecido


````python
import pandas as pd
import numpy as np
from datetime import datetime

print("Calculando métricas pre-calculadas ...")
metricas_ventas = df_items.groupby('product_id').agg(
    price=('price', 'mean'),
    total_units_sold=('order_item_id', 'count')
).reset_index()

# Simulacion de average_rating para cumpli con computed pattern
np.random.seed(42)
metricas_ventas['average_rating'] = np.random.uniform(1.0, 5.0, len(metricas_ventas)).round(1)

# unir metricas
df_catalogo = pd.merge(df_products, metricas_ventas, on='product_id', how='left')

# rellenar nulls
df_catalogo['price'] = df_catalogo['price'].fillna(0).round(2)
df_catalogo['total_units_sold'] = df_catalogo['total_units_sold'].fillna(0).astype(int)
df_catalogo['average_rating'] = df_catalogo['average_rating'].fillna(0)

print(" Estructurando los documentos flexibles (esto puede tomar unos segundos)...")
documentos_productos = []

for index, row in df_catalogo.iterrows():
    categoria_base = str(row['product_category_name']) if pd.notna(row['product_category_name']) else 'general'
    nombre_producto = f"Producto {categoria_base.replace('_', ' ').title()} - {str(row['product_id'])[:5]}"

    specs = {}
    if pd.notna(row['product_name_lenght']):
        specs["name_length"] = str(int(row['product_name_lenght']))
    if pd.notna(row['product_description_lenght']):
        specs["description_length"] = str(int(row['product_description_lenght']))
    if pd.notna(row['product_photos_qty']):
        specs["photos_qty"] = str(int(row['product_photos_qty']))

    doc = {
        "product_id": str(row['product_id']),
        "name": nombre_producto,
        "category": categoria_base,
        "price": float(row['price']),
        "photos": [f"https://ecommify.images/{row['product_id']}_1.jpg"],
        "specifications": specs,
        "computed_metrics": {
            "total_units_sold": int(row['total_units_sold']),
            "average_rating": float(row['average_rating'])
        },
        "dimensions": {
            "weight_g": float(row['product_weight_g']) if pd.notna(row['product_weight_g']) else None,
            "length_cm": float(row['product_length_cm']) if pd.notna(row['product_length_cm']) else None,
            "height_cm": float(row['product_height_cm']) if pd.notna(row['product_height_cm']) else None,
            "width_cm": float(row['product_width_cm']) if pd.notna(row['product_width_cm']) else None
        },
        "timestamps": True,
        "created_at": datetime.utcnow()
    }
    documentos_productos.append(doc)

print(f". Iniciando inyección ...")
coleccion_productos = db['catalogo_enriquecido']

coleccion_productos.delete_many({})


resultado = coleccion_productos.insert_many(documentos_productos)

print(f"¡Éxito total! Se insertaron documentos en la colección 'products'.")
````

### CALCULO DE METRICAS, ARMADO Y INSERCION A BASE DE DATOS 03_seller_metrics\

### Desarrollo y Optimización de Aggregation Pipeline Complejo

````python
import pandas as pd

print("=== DESARROLLO DE AGGREGATION PIPELINE OPTIMIZADO ===")
print("Generando reporte gerencial de ingresos vs. fotos requeridas...")

coleccion_productos = db['catalogo_enriquecido']

pipeline_analitico = [
    # STAGE 1: $match
    {"$match": {"category": {"$in": ["perfumery"]}}},

    # STAGE 2: $sort (
    {"$sort": {"computed_metrics.total_units_sold": -1}},

    # STAGE 3: $project (Proyección Temprana Segura)

    {"$project": {
        "_id": 0,
        "rating": {"$ifNull": ["$computed_metrics.average_rating", 0]},
        "photos": {"$ifNull": ["$photos", []]},
        "ingreso_producto": {
            "$multiply": [
                {"$ifNull": ["$price", 0]},
                {"$ifNull": ["$computed_metrics.total_units_sold", 0]}
            ]
        }
    }},

    # STAGE 4: $unwind
    {"$unwind": {
        "path": "$photos",
        "preserveNullAndEmptyArrays": True
    }},

    # STAGE 5: $group (Agrupación analítica)
    {"$group": {
        "_id": "$rating",
        "total_fotos_cdn": {"$sum": 1},
        "ingreso_promedio": {"$avg": "$ingreso_producto"}
    }},

    # STAGE 6: $sort (Ordenamiento final)
    {"$sort": {"ingreso_promedio": -1}}
]

resultados_pipeline = list(coleccion_productos.aggregate(
    pipeline_analitico,
    allowDiskUse=True
))

df_pipeline = pd.DataFrame(resultados_pipeline)

if not df_pipeline.empty:
    df_pipeline['ingreso_promedio'] = df_pipeline['ingreso_promedio'].apply(lambda x: f"${x:,.2f}")
    df_pipeline.rename(columns={
        '_id': 'Calificación Promedio',
        'total_fotos_cdn': 'Total Fotos en CDN',
        'ingreso_promedio': 'Ingreso Promedio por Producto'
    }, inplace=True)

    display(df_pipeline)
else:
    print("Error está en la estructura de los datos inyectados.")
````
