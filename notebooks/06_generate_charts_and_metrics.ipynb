# 06 - Graficas y metricas
Objetivo: generar graficas antes/despues a partir de los CSV de evidencias.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
base = Path('/content/evidences')
pg = pd.read_csv(base / 'postgresql/postgresql_metrics.csv')
mg = pd.read_csv(base / 'mongodb/mongodb_metrics.csv')

def plot_metric(df, label, out):
    clean = df.dropna(subset=['execution_time_before_ms','execution_time_after_ms'])
    ax = clean.set_index('query')[['execution_time_before_ms','execution_time_after_ms']].plot(kind='bar')
    ax.set_ylabel('ms')
    ax.set_title(label)
    plt.tight_layout()
    plt.savefig(out, dpi=160)

plot_metric(pg, 'PostgreSQL execution time before/after', base / 'postgresql/charts/postgresql_execution_time_before_after.png')
plot_metric(mg, 'MongoDB execution time before/after', base / 'mongodb/charts/mongodb_execution_time_before_after.png')
```
